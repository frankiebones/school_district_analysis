# formatting
> .map( ) _function_  
> .format( ) _function_

```python
def passing_math_percent(math_count, student_count):

    return math_count / float(student_count) * 100
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Students"]
```

```python
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Total Budget"]
```

```python
district_summary_df["Avg Math Score"] = district_summary_df["Avg Math Score"].map("{:.1f}".format)
district_summary_df["Avg Reading Score"] = district_summary_df["Avg Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)

district_summary_df
```

```python
per_school_pass_math = school_complete_df[(school_complete_df["math_score"] >= 70)]
per_school_pass_reading = school_complete_df[(school_complete_df["reading_score"] >= 70)]

print(per_school_pass_math)
print("")
print(per_school_pass_reading)
```

```python
per_school_pass_math = per_school_pass_math.groupby(["School Name"]).count()["student_name"]
per_school_pass_reading = per_school_pass_reading.groupby(["School Name"]).count()["student_name"]

print(per_school_pass_math)
print("")
print(per_school_pass_reading)
```

```python
per_school_pass_math = per_school_pass_math / per_school_counts * 100
per_school_pass_reading = per_school_pass_reading / per_school_counts * 100

print(per_school_pass_math)
print("")
print(per_school_pass_reading)
```
